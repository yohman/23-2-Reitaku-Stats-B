```python
# library import
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import t
import seaborn as sns


# 統計の生徒たち
students = ['Ai','Chikashi','Gunji','honoka','hukuzawa','Junnya','Kairi','Kaneko','Keigo','Koki','Kotaro','Kouta','Koutarou','Naoya','Rento','shigetoshi','shohei','Taiga','Yamato','Yuki','Yuki','Masanobu']

# 正規分布からランダムな整数のテストスコア
test_scores = np.random.normal(loc=80, scale=10, size=len(students)).astype(int)
# test_scores = np.random.randint(60, 101, size=len(students))

# Clip scores above 100
test_scores = np.clip(test_scores, None, 100)

# 生徒の名前とランダムなテストスコアを含むデータフレームを作成する
df = pd.DataFrame({'Name': students, 'Test Score': test_scores})

# 表示
df
```

```python
# Set the style to a different theme
sns.set_style('darkgrid')

# Plot the histogram with a curved line
sns.displot(df['Test Score'], kde=True)

# Show the plot
plt.show()
```

```python
# mean of the test scores
mean_score = df['Test Score'].mean()

# standard deviation of the test scores
std_score = df['Test Score'].std()

var_score = df['Test Score'].var()

print('平均値：',round(mean_score,2),'分散：',round(var_score,2),'標準偏差：',round(std_score,2))
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import t, norm

# Define the degrees of freedom
degrees_of_freedom = [30]

# Generate x values for the plot
x = np.linspace(-4, 4, 100)

# Plot the t-distribution for each degree of freedom
for i, df in enumerate(degrees_of_freedom):
    # Calculate the probability density function (PDF) for the t-distribution
    y = t.pdf(x, df)
    
    # Calculate the alpha value based on the index
    alpha = (i + 1) / len(degrees_of_freedom)
    
    # Get the color from the Blues colormap with adjusted alpha
    color = plt.cm.Blues(alpha)
    
    # Plot the PDF with the adjusted color
    plt.plot(x, y, label=f't-distribution (df={df})', color=color)

# Calculate the probability density function (PDF) for the normal distribution
y_normal = norm.pdf(x, 0, 1)

# Plot the normal distribution as a thicker red line
plt.plot(x, y_normal, label='Normal Distribution', color='red', linewidth=2)

# Add labels and a legend to the plot
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.title('t-Distribution and Normal Distribution')
plt.legend()

# Show the plot
plt.show()
```
