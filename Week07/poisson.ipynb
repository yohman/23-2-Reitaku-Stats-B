```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Define the average rate of occurrence (lambda) for the Poisson distribution
lambda_ = 5

# Generate values from 0 to 15 (for example) and calculate Poisson probabilities
x = np.arange(0, 20)
poisson_probabilities = poisson.pmf(x, lambda_)

# Create a bar chart
plt.bar(x, poisson_probabilities, align='center', alpha=0.7)
plt.xticks(x)
plt.xlabel('Number of Events')
plt.ylabel('Probability')
plt.title('Poisson Distribution (λ=5)')
plt.show()
```

```python
import math

# Define the average rate of occurrence (lambda) for the Poisson distribution
lambda_ = 5

# Define the minimum value of x (4 or more)
min_x = 5

# Define an upper limit for the summation (e.g., 20)
upper_limit = 20

# Calculate the probability of x >= min_x using the Poisson probability mass function
probability = 0
for x in range(min_x, upper_limit + 1):
    prob_x = (math.exp(-lambda_) * lambda_**x) / math.factorial(x)
    probability += prob_x

print(f"The probability of x >= {min_x} with λ={lambda_} is {probability:.4f}")
```
