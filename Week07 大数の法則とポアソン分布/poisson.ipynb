```python
import numpy as np
import matplotlib.pyplot as plt

# Function to roll two dice and calculate their sums
def roll_dice(num_rolls):
    return np.random.randint(1, 7, size=(num_rolls, 2))

# Create the population by rolling two dice 1000 times
population = np.sum(roll_dice(1000), axis=1)

# Plot the distribution of the population
plt.figure(figsize=(10, 5))
plt.hist(population, bins=50, range=(2, 13), density=True, alpha=0.6)
plt.title('Population Distribution of Two Dice Rolls')
plt.xlabel('Sum of Two Dice')
plt.ylabel('Probability')
plt.show()

# Draw random samples of different sizes and plot their distributions
sample_sizes = [5, 10, 20, 30]

plt.figure(figsize=(12, 8))
for i, size in enumerate(sample_sizes):
    plt.subplot(2, 2, i + 1)
    means = [np.mean(np.sum(roll_dice(size), axis=1)) for _ in range(1000)]
    plt.hist(means, bins=30, density=True, alpha=0.6)
    plt.title(f'Sample Size: {size}')
    plt.xlabel('Mean of Two Dice Rolls')
    plt.ylabel('Probability')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Define the average rate of occurrence (lambda) for the Poisson distribution
lambda_ = 3

# Generate values from 0 to 15 (for example) and calculate Poisson probabilities
x = np.arange(0, 20)
poisson_probabilities = poisson.pmf(x, lambda_)

# Create a bar chart
plt.bar(x, poisson_probabilities, align='center', alpha=0.7)
plt.xticks(x)
plt.xlabel('Number of Events')
plt.ylabel('Probability')
plt.title('Poisson Distribution (λ='+str(lambda_)+')')
plt.show()
```

```python
import math

# Define the average rate of occurrence (lambda) for the Poisson distribution
lambda_ = 5

# Define the minimum value of x (4 or more)
min_x = 5

# Define an upper limit for the summation (e.g., 20)
upper_limit = 20

# Calculate the probability of x >= min_x using the Poisson probability mass function
probability = 0
for x in range(min_x, upper_limit + 1):
    prob_x = (math.exp(-lambda_) * lambda_**x) / math.factorial(x)
    probability += prob_x

print(f"The probability of x >= {min_x} with λ={lambda_} is {probability:.4f}")
```
