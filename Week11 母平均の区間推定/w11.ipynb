```python
import numpy as np
from scipy.stats import ttest_1samp
import matplotlib.pyplot as plt

# Set seed for reproducibility
np.random.seed(42)

# Generate random test scores from a normal distribution
test_scores = np.random.normal(80, 10, 20)

# Perform one-sample t-test
t_statistic, p_value = ttest_1samp(test_scores, 80)

# Define the significance level
alpha = 0.05

# Print the null hypothesis
print("Null Hypothesis: The population average is 80")

# Plot the distribution of test scores
plt.hist(test_scores, bins=10, edgecolor='black')
plt.xlabel('Test Scores')
plt.ylabel('Frequency')
plt.title('Distribution of Test Scores')
plt.axvline(x=80, color='r', linestyle='--', label='Population Average: 80')
plt.axvline(x=np.mean(test_scores), color='g', linestyle='-', label='Sample Average: {:.2f}'.format(np.mean(test_scores)))
plt.legend()
plt.show()

# Plot the t-distribution cutoff
df = len(test_scores) - 1
x = np.linspace(-4, 4, 100)
t_values = t.pdf(x, df)
plt.plot(x, t_values, 'r-', lw=2)
plt.axvline(x=t.ppf(1 - alpha/2, df), color='b', linestyle='--', label='T-Distribution Cutoff')
plt.xlabel('T-Statistic')
plt.ylabel('Probability Density')
plt.title('T-Distribution')
plt.legend()
plt.show()
```

```python
import numpy as np
import seaborn as sns
import ipywidgets as widgets
from IPython.display import display
import matplotlib.pyplot as plt

# Set seed for reproducibility
np.random.seed(42)

# Parameters for the normal distribution
mu = 1000
std_dev = 100

# Generate random data from a normal distribution
data = np.random.normal(mu, std_dev, 1000)

# Function to update the plot
def update_plot(change):
    # Get the current value of n
    n = change.new
    
    # Calculate the confidence interval
    confidence_level = 0.95
    z_score = 1.96  # For a 95% confidence interval
    sample = np.random.choice(data, size=n, replace=False)
    sample_mean = np.mean(sample)
    sample_std_dev = np.std(sample)
    margin_of_error = z_score * (sample_std_dev / np.sqrt(n))
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error
    
    # Plot the data and confidence interval using seaborn
    plt.figure(figsize=(8, 6))
    sns.histplot(data, bins=20, kde=True)
    plt.axvline(lower_bound, color='r', linestyle='--', label='Lower Bound: {:.2f}'.format(lower_bound))
    plt.axvline(upper_bound, color='r', linestyle='--', label='Upper Bound: {:.2f}'.format(upper_bound))
    plt.axvline(mu, color='g', linestyle='-', label='Population Mean: {:.2f}'.format(mu))
    plt.legend()
    plt.xlabel('Data')
    plt.ylabel('Frequency')
    plt.title('Confidence Interval (n={}, alpha={:.2f})'.format(n, 1 - confidence_level))
    plt.show()

# Create a dropdown widget for changing the value of n
n_widget = widgets.IntSlider(value=5, min=1, max=100, step=1, description='n:')
n_widget.observe(update_plot, 'value')

# Display the dropdown widget
display(n_widget)
```

```python
import numpy as np
import ipywidgets as widgets
from IPython.display import display
import plotly.graph_objects as go

# Set seed for reproducibility
np.random.seed(42)

# Parameters for the normal distribution
mu = 1000
std_dev = 100

# Generate random data from a normal distribution
data = np.random.normal(mu, std_dev, 1000)

# Function to update the plot
def update_plot(change):
    # Get the current value of n
    n = change.new
    
    # Calculate the confidence interval
    confidence_level = 0.95
    z_score = 1.96  # For a 95% confidence interval
    sample = np.random.choice(data, size=n, replace=False)
    sample_mean = np.mean(sample)
    sample_std_dev = np.std(sample)
    margin_of_error = z_score * (sample_std_dev / np.sqrt(n))
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error
    
    # Create the figure
    fig = go.Figure()
    
    # Add the histogram trace
    fig.add_trace(go.Histogram(x=data, nbinsx=20, histnorm='probability', name='Data'))
    
    # Add the vertical lines for the confidence interval
    fig.add_shape(type='line', x0=lower_bound, x1=lower_bound, y0=0, y1=1, line=dict(color='red', dash='dash'), name='Lower Bound')
    fig.add_shape(type='line', x0=upper_bound, x1=upper_bound, y0=0, y1=1, line=dict(color='red', dash='dash'), name='Upper Bound')
    
    # Add the text labels for the confidence interval
    fig.add_annotation(x=lower_bound, y=0.5, text=f'Lower Bound: {lower_bound:.2f}', showarrow=False, font=dict(color='red'))
    fig.add_annotation(x=upper_bound, y=0.5, text=f'Upper Bound: {upper_bound:.2f}', showarrow=False, font=dict(color='red'))
    
    # Add the population mean line
    fig.add_shape(type='line', x0=mu, x1=mu, y0=0, y1=1, line=dict(color='green'), name='Population Mean')
    
    # Update the layout
    fig.update_layout(
        title=f'Confidence Interval (n={n}, alpha={1 - confidence_level:.2f})',
        xaxis_title='Data',
        yaxis_title='Probability',
        barmode='overlay',
        bargap=0.1
    )
    
    # Show the figure
    fig.show()

# Create a dropdown widget for changing the value of n
n_widget = widgets.IntSlider(value=5, min=1, max=100, step=1, description='n:')
n_widget.observe(update_plot, 'value')

# Display the dropdown widget
display(n_widget)
```
